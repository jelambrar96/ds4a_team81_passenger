```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Leer dataframe grande
df1 = pd.read_csv("./data/Operacion_de_pasajeros_y_despacho_de_veh_culos_en_la_modalidad_de_transporte_de_pasajeros_por_carretera.csv")
```

```python
#Crear columna Ruta concatenando los códigos de los municipios origen con municipios destino
df1["Ruta"]=df1["MUNICIPIO_ORIGEN_RUTA"].astype(str)+"-"+df1["MUNICIPIO_DESTINO_RUTA"].astype(str)
```

```python
#Conocer las rutas más frecuentes
df1.groupby("Ruta").count()[["TERMINAL"]].sort_values(by="TERMINAL",ascending=False)
```

```python
#Crear columna fecha de despacho 2 quitando la hora de la columna fecha de despacho 1
df1["FECHA_DESPACHO 2"]=df1["FECHA_DESPACHO"].apply(lambda x:x[:10])
```

```python
#Colocar formato MES DIA AÑO 
df1["FECHA_DESPACHO 2"]=pd.to_datetime(df1["FECHA_DESPACHO 2"])
```

```python
#Crear marca de tiempo para conocer fechas mayores a esta que pueden ser erróneos
date=pd.Timestamp('2021-05-01 00:00:00')
```

```python
#Imprimir filas con fecha mayor a la marca de tiempo anterior
df1[df1["FECHA_DESPACHO 2"]>date]
```

```python
df1["PASAJEROS"]=df1["PASAJEROS"].astype(str).str.replace(",","").astype(int)
```

```python
df3 = df1.groupby(by = ['NIVEL_SERVICIO','FECHA_DESPACHO 2'])[['PASAJEROS']].sum().reset_index()
df3['FECHA_DESPACHO 2'] = pd.to_datetime(df3['FECHA_DESPACHO 2']).apply(lambda x: x.strftime('%Y/%m'))
```

```python
df3.groupby(by = 'NIVEL_SERVICIO')[['PASAJEROS']].sum()
```

```python
fig, ax = plt.subplots(figsize=(20,10))
ax.set_title('Passengers per month on each service level')
sns.lineplot(data = df3, x = 'FECHA_DESPACHO 2', y = 'PASAJEROS', hue = 'NIVEL_SERVICIO')
```

```python
df4 = df1.groupby(by = ['NIVEL_SERVICIO','MUNICIPIO_ORIGEN_RUTA'])[['PASAJEROS']].sum().reset_index()
df4 = df4.pivot("NIVEL_SERVICIO", "MUNICIPIO_ORIGEN_RUTA", "PASAJEROS").fillna(0)
df4
```

```python
ax = sns.heatmap(df4)
```
